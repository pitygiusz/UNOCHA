```python
import polars as pl

# Read with Spark, convert to pandas, then to Polars
hrp_spark = spark.table("unocha.default.humanitarian_response_plans")
hn_spark = spark.table("unocha.default.hpc_hn")
cod_spark = spark.table("unocha.default.cod_population")
hrp_pol = pl.from_pandas(hrp_spark.toPandas())
cod_pol = pl.from_pandas(cod_spark.toPandas())
hn_pol = pl.from_pandas(hn_spark.toPandas())

# Now use Polars operations
# df_polars.head()
# df_polars.describe()
```

```python
pop = spark.table("unocha.default.years_pop")
```

```python
import pandas as pd

pop_pd = pop.toPandas()
pop_pd.head()
```

```python
pop_corr = pop_pd.melt(id_vars=["Country Code"], 
                  var_name="year", 
                  value_name="population")
```

```python
pop_corr = pop_corr.rename(columns={"Country Code": "Country"},)
```

```python
pop_corr.head()
```

```python
pop_sp = spark.createDataFrame(pop_corr)
## Write Frame out as Table
pop_sp.write.mode("overwrite").saveAsTable("unocha.default.population_full")
```
