```sql
CREATE OR REPLACE TABLE unocha.default.fts_requirements_funding_global_cleaned AS
SELECT
  countryCode AS country_code,
  name AS requirement_name,
  typeName as requirement_type_name,
  startDate AS start_date,
  endDate AS end_date,
  requirements AS required_funds,
  funding AS granted_funds,
  percentFunded AS granted_percentage
FROM unocha.default.fts_requirements_funding_global
WHERE requirements IS NOT NULL OR funding IS NOT NULL
```
