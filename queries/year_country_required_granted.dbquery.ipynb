```sql
create or replace table unocha.default.year_country_required_granted as
select
  year(start_date) as start_year,
  country_code,
  sum(required_funds) as total_required_funds,
  sum(granted_funds) as total_granted_funds,
  round(sum(granted_funds) / (sum(required_funds)+1) * 100, 2) as total_granted_percentage

from unocha.default.fts_requirements_funding_global_cleaned

where start_date is not null

group by country_code, year(start_date)
```
