```sql
CREATE OR REPLACE TABLE unocha.default.master_table AS
with sev_inds as (select country, Year, max(SevIn) as severity_index
from unocha.default.sev_in
group by country, Year)
select
  mas.*,
  sev_inds.severity_index,
  codes.sector as sector
from unocha.default.master_table_step_3 as mas
left join sev_inds
  on mas.country_code = sev_inds.country and mas.year = sev_inds.Year
left join unocha.default.cluster_code_sector as codes
  on mas.cluster_code = codes.cluster_code
order by year desc, country_code
```
