```sql
CREATE OR REPLACE TABLE unocha.default.master_table_step_3 AS
select
  mas.*,
  rg.total_required_funds,
  rg.total_granted_funds,
  rg.total_granted_percentage
from unocha.default.master_table_step_2 mas
full join unocha.default.year_country_required_granted rg
  on mas.country_code = rg.country_code and mas.year = rg.start_year
```
