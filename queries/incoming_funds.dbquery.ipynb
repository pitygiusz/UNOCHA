```sql
CREATE OR REPLACE TABLE unocha.default.fts_incoming_funding_cleaned AS
SELECT 
  date AS grant_date,
  description AS grant_description,
  amountUSD AS granted_funds,
  srcOrganization AS source_organization,
  srcOrganizationTypes AS source_organization_type,
  srcLocations AS source_country_code,
  destPlan AS requirement_name,
  destOrganization AS destination_origanization,
  destOrganizationTypes AS destination_organization_type,
  destGlobalClusters AS destination_clusters,
  destLocations AS destination_country_codes,
  status AS status
FROM unocha.default.fts_incoming_funding_global
WHERE amountUSD IS NOT NULL
```
