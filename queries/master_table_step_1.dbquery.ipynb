```sql
CREATE OR REPLACE TABLE unocha.default.master_table_step_1 AS
select
  hpc.`Country ISO3` as country_code,
  hpc.Cluster as cluster_code,
  hpc_copy.Population as population,
  hpc.`In Need` as in_need,
  hpc.Targeted as targeted,
  hpc.Year as year
from unocha.default.hpc_hno as hpc
full join (select * from unocha.default.hpc_hno where Population is not null and Category is null) as hpc_copy
  on hpc.`Country ISO3` == hpc_copy.`Country ISO3`
  and hpc.Year == hpc_copy.Year
where hpc.`Admin 1 PCode` is null
and hpc.`Admin 2 PCode` is null
and hpc.`Admin 3 PCode` is null
and hpc.Category is null
```
